Задачи исследования:

* посчитать, сколько книг вышло после 1 января 2000 года;
* для каждой книги посчитать количество обзоров и среднюю оценку;

## Посчитать, сколько книг вышло после 1 января 2000 года

```python
count_books = """ SELECT COUNT(book_id)
                  FROM books
                  WHERE publication_date > CAST('2000-01-01' as date)
              """
output(count_books)
```

После 1 января 2000 года вышло 819 книг

## Для каждой книги посчитать количество обзоров и среднюю оценку

```python
count_reviews_and_means = """ WITH
                              rev AS
                              (SELECT book_id, 
                                      COUNT(review_id) OVER (PARTITION BY book_id) AS count
                              FROM reviews),
                              rat AS 
                              (SELECT book_id, 
                                      ROUND(AVG(rating) OVER (PARTITION BY book_id), 2) AS avg_rating
                              FROM ratings) 

                              SELECT DISTINCT(b.book_id),
                                     b.title, 
                                     count, 
                                     avg_rating
                              FROM books AS b
                              LEFT JOIN rev ON rev.book_id = b.book_id
                              LEFT JOIN rat ON rat.book_id = b.book_id
                              ORDER BY 1 
                              
                          """
output(count_reviews_and_means)
```

Посчитано количество обзоров и средняя оценка для каждой книги

## Определить издательство, которое выпустило наибольшее число книг толще 50 страниц

```python
publisher = """ SELECT publishers.publisher,
                      COUNT(books.book_id) AS count
                FROM publishers
                INNER JOIN books ON books.publisher_id = publishers.publisher_id
                WHERE books.num_pages > 50
                GROUP BY 1
                ORDER BY 2 DESC
                LIMIT 10
            """

output(publisher)
```

Издательство Penguin Books выпустило наибольшее число книг толще 50 страниц

## Определить автора с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками

```python
author_max = """ WITH rat AS (SELECT book_id, 
                                    COUNT(rating_id) 
                             FROM ratings
                             GROUP BY 1
                             HAVING COUNT(rating_id) > 50)
            
                SELECT a.author,
                       AVG(r.rating) 
                FROM books AS b
                INNER JOIN ratings AS r ON r.book_id = b.book_id 
                INNER JOIN authors AS a ON a.author_id = b.author_id
                WHERE b.book_id IN (SELECT book_id FROM rat)
                GROUP BY 1
                ORDER BY 2 DESC
                
                LIMIT 1
            """

output(author_max)

```

Автор с самой высокой средней оценкой книг — учитывать только книги с 50 и более оценками - J.K. Rowling

## Посчитать среднее количество обзоров от пользователей, которые поставили больше 48 оценок

```python
review_avg = """ SELECT AVG(review.review_cnt) AS avg_review
                          FROM (SELECT COUNT(review_id) AS review_cnt
                                FROM reviews
                                WHERE username IN (SELECT username
                                                   FROM ratings
                                                   GROUP BY 1
                                                   HAVING COUNT(ratings) > 48)
                               GROUP BY username) AS review; 
             """
output(review_avg)
```

24 - среднее количество обзоров от пользователей, которые поставили больше 48 оценок
